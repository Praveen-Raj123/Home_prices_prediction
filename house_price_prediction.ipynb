```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("homeprices.xlsx") #need to use read_excel instead of read_csv because it is xlsx format
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue')
plt.show()
```

```python
data = pd.read_excel('area.xlsx')
data.head(3)
```

```python
new_data = reg.predict(data)
```

```python
data['prices'] = new_data
```

```python
data.to_csv("predicted.csv")
```
